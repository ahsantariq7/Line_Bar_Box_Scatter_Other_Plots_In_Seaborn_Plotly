```python
import seaborn as sns
import matplotlib.pyplot as plt 

phool=sns.load_dataset("iris")
phool
```

```python
sns.barplot(x="species",y="sepal_width",data=phool)
plt.title("Iris DataSet")
plt.figure(figsize=(10,9))
sns.set_style("whitegrid")

plt.show()
```

```python
sns.barplot(x="species",y="petal_width",data=phool)
plt.title("Iris DataSet")
plt.figure(figsize=(10,9))
sns.set_style("whitegrid")

plt.show()
```

```python
sns.barplot(x="species",y="petal_length",data=phool)
plt.title("Iris DataSet")
plt.figure(figsize=(10,9))
sns.set_style("whitegrid")

plt.show()
```

```python
titanic_data=sns.load_dataset("titanic")
titanic_data
```

```python
sns.barplot(x="who",y="alone",hue="sex",data=titanic_data)
plt.title("Iris DataSet")
plt.figure(figsize=(10,9))
sns.set_style("dark")

plt.show()
```

```python
sns.barplot(x="sex",y="alone",hue="who",data=titanic_data)
plt.title("Iris DataSet")
plt.figure(figsize=(10,9))
sns.set_style("dark")

plt.show()
```

```python
sns.barplot(x="sex",y="alone",hue="who",data=titanic_data,order=["child","female","male"])
plt.title("Iris DataSet")
plt.figure(figsize=(10,9))
sns.set_style("dark")

plt.show()
```
